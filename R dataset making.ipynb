## All csv files

```python
dirlist=["D:\dane\dane1 - B00702A", "D:\dane\dane2 - B00608S", "D:\dane\dane3 - B00305A"]
```

```python
for dir in dirlist:
    os.chdir(str(dir))
    lista = os.listdir(dir)
    
    names=["stacja", 1, "data", "wartość", 4]
    df=pd.DataFrame()

    for plik in lista:
        print(plik)
        dane = pd.read_csv(plik, header=None,sep=';', names=names)
        dane=dane.drop([1,4], axis=1)
        dane = dane.drop_duplicates(subset=['data', 'stacja'])
        dane=dane.pivot(index='data', columns='stacja', values='wartość')
        if "253230160" in dane.keys():
            dane=dane.rename(columns={'253230160': 253230160})
        if "249190240" in dane.keys():
            dane=dane.rename(columns={'249190240': 249190240})
        if "249190440" in dane.keys():
            dane=dane.rename(columns={'249190440': 249190440})
        dane = dane[[253230160, 249190240, 249190440]]
        for column in dane:
            dane[column]=dane[column].str.replace(',', '.').astype(float)
        df=pd.concat([df, dane])
    df.to_csv(str(str(dir) + ' - outv2.csv'))
```
